```python
import os.path
from psyrun.io import load_dict_h5
%matplotlib inline
```

```python
data = pd.DataFrame(load_dict_h5(os.path.join('psywork', 'n_neuron_scaling_by_d', 'result.h5')))
```

```python
sns.set('paper', style='ticks', font_scale=1.5)
```

```python
g = sns.lmplot(
    'd', 'e_noise', data, hue='scaling', col='encoders',
    hue_order=['linear', 'quadratic'],
    col_order=['uniform', 'scaled orthonormal'],
    logx=True, sharex=False, sharey=False, legend_out=False)
g.set_axis_labels('Dimensions $D$', 'Squared error $E^2$')
titles = ['Single ensemble', 'One ensemble for each dimension']
for ax, title in zip(g.axes.flat, titles):
    ax.set_title(title)
    ax.set_xlim(left=0.)
    ax.set_ylim(bottom=0.)
    if ax.get_xlim()[1] > 150:
        ax.set_xlim(right=150)
    if ax.get_ylim()[1] > 1.:
        ax.set_ylim(top=1.)
plt.tight_layout()
```
